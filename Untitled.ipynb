```python
np_description = df['Description'].unique()
i = 0
for i in range(df['Description'].unique().size):
    df.replace(np_description[i],float(i),inplace = True)
    print(i)
```

```python
np_description = df['Country'].unique()
i = 0
for i in range(df['Country'].unique().size):
    df.replace(np_description[i],float(i),inplace = True)
```

```python
df = df.drop(['InvoiceNo','StockCode','InvoiceDate'], axis=1)
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,cmap=sns.cm.rocket_r)
```

```python
from sklearn.cluster import KMeans
from sklearn.pipeline import make_pipeline
inertias = []
ks = range(1, 15)
for i in ks:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    inertias.append(kmeans.inertia_)
plt.plot(ks, inertias, '-o')
plt.xlabel('Nmero de clusters')
plt.ylabel('Inrcia')
```

```python
inertias = []
ks = range(1, 20)
for i in ks:
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df)
    inertias.append(kmeans.inertia_)
    print(i)
plt.plot(ks, inertias, '-o')
plt.xlabel('Nmero de clusters')
plt.ylabel('Inrcia')
```
